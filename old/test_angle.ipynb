```python
def random_room_dimensions():
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    return np.array([x, y])

def random_head_position(room_dim):
    rdx, rdy = room_dim
    return 0.5*room_dim

def random_mouth_position(head_pos, head_ang):
    hpx, hpy = head_pos
    rdx = random.uniform(-0.01, 0.01)
    rdy = random.uniform(0.11, 0.15)
    rdz = random.uniform(0.02, 0.04)
    x = hpx + rdx*np.cos(head_ang) - rdy*np.sin(head_ang)
    y = hpy + rdx*np.sin(head_ang) + rdy*np.cos(head_ang)
    return np.array([x, y])

def random_ears_position(head_pos, head_ang):
    hpx, hpy = head_pos
    rdx = random.uniform(0.08, 0.09)
    ear_center_l = np.array([hpx - rdx*np.cos(head_ang), hpy - rdx*np.sin(head_ang)])
    ear_center_r = np.array([hpx + rdx*np.cos(head_ang), hpy + rdx*np.sin(head_ang)])
    return np.array([ear_center_l, ear_center_r])
```

```python
room_dim = random_room_dimensions()
print(room_dim)
```

```python
head_position = random_head_position(room_dim)
print(head_position)
```

```python
angle = 0
```

```python
mouth_position = random_mouth_position(head_position, angle)
print(mouth_position)
```

```python
ears_position = random_ears_position(head_position, angle)
print(ears_position)
```

```python
room = pra.ShoeBox(room_dim, fs=16000)
room.add_source(mouth_position)
room.add_microphone_array(ears_position.T)
room.plot()
```
